```python
import igraph as ig
import leidenalg as la
from sklearn.neighbors import NearestNeighbors

def cluster_knn_graph_with_hamming(data_matrix, n_neighbors=15, resolution=0.1, random_seed=12):
    """
    Computes a KNN graph using BallTree and Hamming distance, then applies 
    Leiden clustering to the resulting graph.

    Args:
        data_matrix (np.ndarray): The input data matrix (n samples, m features). 
                                  Must be categorical/binary for Hamming distance.
        n_neighbors (int): The number of neighbors 'k' for the KNN graph.
        resolution (float): The resolution parameter for the Leiden algorithm.
        random_seed (int): Seed for reproducibility.

    Returns:
        tuple: (list of cluster assignments, igraph.Graph object)
    """
    # 1. Validate Input Data for Hamming Distance
    # Hamming distance works best on integers (0/1 or categorical IDs)
    if not np.issubdtype(data_matrix.dtype, np.integer):
        print("Warning: Data matrix is not of integer type. Hamming distance is typically used for binary or categorical (encoded) data.")
    

    knn_model = NearestNeighbors(
        n_neighbors=n_neighbors, 
        algorithm='ball_tree', 
        metric='hamming'
    )
    
    knn_model.fit(data_matrix)
    
    distances, indices = knn_model.kneighbors(data_matrix, return_distance=True)
    
    n_samples = data_matrix.shape[0]

    # 3. Convert the KNN results to an igraph Graph
    
    # Prepare edge list (source, target, weight)
    # The 'weight' is often the distance, but for unweighted clustering, 
    # we can just use the indices.
    
    edge_list = []
    
    # Iterate over each sample
    for i in range(n_samples):
        # Iterate over its neighbors (j > i to avoid duplicate edges in an undirected graph)
        # Note: Scikit-learn includes the point itself as the first neighbor (distance 0), 
        # so we typically start from index 1.
        for j in range(1, n_neighbors): 
            neighbor_index = indices[i, j]
            distance = distances[i, j]
            
            # Create an edge. We use the distance as the edge weight.
            # Only add the edge if the indices are distinct (i != neighbor_index)
            if i != neighbor_index:
                edge_list.append((i, neighbor_index, distance))

    # Create the igraph Graph object
    G = ig.Graph.TupleList(edge_list, directed=False, weights=True)

    # Note: igraph nodes are zero-indexed, which matches the sample indices.
    # The graph may contain duplicate edges if (i, j) and (j, i) were both added.
    # Simplifying the graph to handle this.
    G.simplify(combine_edges='min')
    print(f"Graph created with {G.vcount()} vertices and {G.ecount()} edges.")
    
    # 4. Apply Leiden Clustering (Community Detection)
    print("Applying Leiden clustering...")
    
    # Set the random seed for reproducibility
    np.random.seed(random_seed) 
    
    partition = la.CPMVertexPartition(G, resolution_parameter = 0.1)
    # Use the quality function ModularityVertexPartition (common for unweighted/weighted graphs)
    partition = la.find_partition(
        G, 
        la.CPMVertexPartition, 
        weights='weight', # Use the distances as weights (e.g., closer = stronger connection)
        seed=random_seed,
    )
    
    # The membership is the cluster assignment for each sample
    cluster_assignments = partition.membership
    num_clusters = len(partition)

    print(f"Leiden clustering completed. Found {num_clusters} communities.")

    return cluster_assignments, G



```
